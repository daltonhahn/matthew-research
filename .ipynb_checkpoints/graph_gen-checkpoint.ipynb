```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import math
from statistics import mean, stdev
```

```python
def gen_graph(num_nodes, deg_max):
    deg_max = num_nodes-1
    din = [random.randint(1,deg_max) for _ in range(num_nodes)]
    dout = din
    random.shuffle(dout)
    if nx.is_graphical(dout):
        G = nx.directed_configuration_model(din, dout, seed=42)
        G = nx.DiGraph(G)
        if nx.is_weakly_connected(G):
            return G
        else:
            gen_graph(num_nodes, deg_max)
    else:
        gen_graph(num_nodes, deg_max)
```

```python
def clean_edges(graph):
    output, seen = [], set()
    for item in graph.edges():
        t1 = tuple(item)
        if t1 not in seen and tuple(reversed(item)) not in seen:
            seen.add(t1)
            output.append(item)
    rem_edges = [(v,u) for v,u in graph.edges() if (graph.has_edge(v,u) and graph.has_edge(u,v))]
    D = nx.DiGraph()
    D.add_nodes_from(graph.nodes())
    D.add_edges_from(output)
    return D
```

```python
def get_new_name(pokemon_names, running_list):
    name = pokemon_names[random.randint(0,len(pokemon_names)-1)]
    if name not in running_list:
        return name
    else:
        return get_new_name(pokemon_names, running_list)
```

```python
def add_names(graph):
    pok_df = pd.read_csv('pokemon.csv', index_col=0)
    pok_names = pok_df['Name'].tolist()
    name_map = dict()
    node_names = list()
    for n in graph.nodes():
        name = get_new_name(pok_names, node_names)
        node_names.append(name)
        name_map[n] = name

    G = nx.relabel_nodes(graph, name_map)
    return G
```

```python
G = gen_graph(6, 3)
G = clean_edges(G)
G = add_names(G)
pos = nx.spring_layout(G)
nx.draw(G,pos, with_labels=True)
plt.show()
```
